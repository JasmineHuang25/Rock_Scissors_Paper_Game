## Reload the Saved Model

```python
model = keras.models.load_model('./OtherNotebook_and_Files/Saved_model.h5')
```

## Testing Prediction

```python
path = "./Data_and_img/rock.png"

img = utils.load_img(path,
                     target_size=(150, 150),
                     color_mode="rgb")

input_arr = utils.img_to_array(img)
input_arr = np.array([input_arr])

classes = model.predict(input_arr, batch_size=10)


for i in classes:
  if i[0] == 1:
    print("paper")
  elif i[1] == 1:
    print("Rock")
  else:
    print("Scissor")
        
plt.imshow(img)
print(classes)
```
